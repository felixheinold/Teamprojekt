```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    extracted_text = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Versuche, den Haupttextbereich zu identifizieren
            text = page.filter(lambda obj: obj["object_type"] == "char").extract_text()
            if text:
                # Optional: FuÃŸzeilen und Kopfzeilen rausfiltern
                lines = text.split('\n')
                filtered_lines = [
                    line for line in lines
                    if not is_footer_or_header(line)
                ]
                cleaned_text = "\n".join(filtered_lines)
                extracted_text.append(cleaned_text)
    
    return "\n\n".join(extracted_text)

def is_footer_or_header(line):
    # Passe diese Funktion ggf. an dein PDF-Layout an
    line = line.strip()
    if len(line) < 10 and line.isdigit():
        return True  # Seitenzahlen
    if "Seite" in line or "Page" in line:
        return True
    if "Â©" in line or "www." in line:
        return True
    return False

# Beispielnutzung
pdf_datei = "Marketing_WS_22_23_Kapitel_6-ohneinhalt.pdf" # Ersetze durch den Pfad zu deiner PDF-Datei
text = extract_text_from_pdf(pdf_datei)
print(text)
```

```python
# Nur Nomen (Substantive) extrahieren
def extract_nouns(text):
    doc = nlp(text)
    return [token.text for token in doc if token.pos_ == "NOUN"]
```

```python
def rank_nouns_with_keybert(text, nouns, top_n=5):
    keywords = keybert_model.extract_keywords(
        text,
        keyphrase_ngram_range=(1, 1),
        stop_words=None,
        top_n=top_n * 3
    )
    keyword_list = [kw[0] for kw in keywords]
    noun_set = set(noun.lower() for noun in nouns)

    # Schnittmenge mit lower-case Vergleich
    ranked = []
    for word in keyword_list:
        if word.lower() in noun_set and word not in ranked:
            ranked.append(word)
        if len(ranked) == top_n:
            break

    return ranked
```

```python
nlp = spacy.load("de_core_news_sm")

def generate_noun_gap_text(text, nouns):
    blanks = {}
    used_nouns = set()
    used_sentences = set()
    placeholder_index = 1

    doc = nlp(text)
    new_text = text

    for sent in doc.sents:
        sent_text = sent.text
        replaced = False

        for noun in nouns:
            if noun.lower() in used_nouns:
                continue

            # Regex fÃ¼r exakte Wortgrenze, case-insensitive
            pattern = r'\b' + re.escape(noun) + r'\b'
            match = re.search(pattern, sent_text, flags=re.IGNORECASE)

            if match:
                placeholder = f"___({placeholder_index})___"
                replaced_sent = re.sub(pattern, placeholder, sent_text, count=1, flags=re.IGNORECASE)

                # Ersetze den Satz im Gesamttext
                new_text = new_text.replace(sent_text, replaced_sent, 1)

                blanks[placeholder] = match.group(0)
                used_nouns.add(noun.lower())
                used_sentences.add(sent_text)
                placeholder_index += 1
                replaced = True
                break  # Nur eine LÃ¼cke pro Satz

        if placeholder_index > len(nouns):
            break  # Keine weiteren LÃ¼cken nÃ¶tig

    return new_text, blanks
```

```python
def split_text_into_sections(text, min_length=200):
    raw_sections = [s.strip() for s in text.split('\n\n') if s.strip()]
    sections = []
    buffer = ""

    for part in raw_sections:
        if len(buffer) + len(part) < min_length:
            buffer += " " + part
        else:
            if buffer:
                sections.append(buffer.strip())
            buffer = part
    if buffer:
        sections.append(buffer.strip())

    return sections

def save_all_exercises(filename, data):
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
def process_text_with_nouns(input_text, output_filename="lueckentext_nomen.json", gaps_per_section=5):
    sections = split_text_into_sections(input_text)
    exercises = []

    for i, section in enumerate(sections):
        print(f"\nğŸ§© Abschnitt {i + 1}")
        print(f"ğŸ“„ Originaltext:\n{section}\n")

        all_nouns = extract_nouns(section)
        print(f"ğŸ” Gefundene Nomen ({len(all_nouns)}): {all_nouns}")

        ranked_nouns = rank_nouns_with_keybert(section, all_nouns, top_n=gaps_per_section)
        print(f"â­ï¸ AusgewÃ¤hlte LÃ¼ckenwÃ¶rter: {ranked_nouns}")

        gap_text, answers = generate_noun_gap_text(section, ranked_nouns)
        print(f"\nâœï¸ LÃ¼ckentext:\n{gap_text}")
        print(f"âœ… LÃ¶sungen: {answers}\n")

        exercises.append({
            "section_id": i + 1,
            "original_text": section,
            "gap_text": gap_text,
            "answers": answers
        })

    save_all_exercises(output_filename, exercises)
    print(f"\nâœ… {len(exercises)} LÃ¼ckentexte gespeichert in '{output_filename}'.")

```
